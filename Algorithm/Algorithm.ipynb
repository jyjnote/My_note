## **Insertion-sort**
### This algorithm is a basic algorithm. It stores the a[key] in an array starting from [j to n] and changes the position if it is less than a[i].
<br/><br/>
<div style="border: 2px solid green; padding: 5px; background-color: yellow; border-radius: 5px; color: green; display: inline-block;">
     pseudo-code
</div>

 인설트_솔팅(배열 arr, 정수 n):
     반복 for i = 1 to n - 1:
        현재값 = arr[i]
        이전위치 = i - 1

        while 이전위치 >= 0 and arr[이전위치] > 현재값:
            arr[이전위치 + 1] = arr[이전위치]
            이전위치 = 이전위치 - 1

        arr[이전위치 + 1] = 현재값

```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        key = arr[i]  # 현재 요소를 선택하여 삽입할 위치 찾기
        j = i - 1

        # key보다 큰 원소들을 오른쪽으로 이동
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key  # key를 정렬된 부분에 삽입

# 예시
arr = [12, 11, 13, 5, 6]
print("정렬 전 배열:", arr)

insertion_sort(arr)
print("정렬 후 배열:", arr)
```

## **Merge-sort**
### The following algorithm operates based on a recursive algorithm. This is integrated under the assumption that the two arrays are aligned. The principle is to compare values one by one from two sorted arrays and put them into a new array. The time complexity of this algorithm is Θ(n log n).
<br/><br/><br/><br/>
<div style="border: 2px solid green; padding: 5px; background-color: yellow; border-radius: 5px; color: green; display: inline-block;">
     pseudo-code
</div>

    머지_정렬(배열 arr):
        만약 배열의 길이가 1 이하이면, 이미 정렬되어 있다고 판단하고 반환

        중간지점 = 배열의 길이 // 2
        왼쪽_배열 = arr의 첫부분부터 중간지점까지의 요소
        오른쪽_배열 = arr의 중간지점부터 끝까지의 요소

        머지_정렬(왼쪽_배열)  # 왼쪽 부분 배열에 대해 재귀 호출
        머지_정렬(오른쪽_배열)  # 오른쪽 부분 배열에 대해 재귀 호출

        왼쪽_인덱스 = 0
        오른쪽_인덱스 = 0
        병합된_배열의_인덱스 = 0

        # 두 부분 배열을 병합
        while 왼쪽_인덱스 < 왼쪽_배열의_길이 and 오른쪽_인덱스 < 오른쪽_배열의_길이:
            if 왼쪽_배열[왼쪽_인덱스] <= 오른쪽_배열[오른쪽_인덱스]:
                arr[병합된_배열의_인덱스] = 왼쪽_배열[왼쪽_인덱스]
                왼쪽_인덱스 = 왼쪽_인덱스 + 1
            else:
                arr[병합된_배열의_인덱스] = 오른쪽_배열[오른쪽_인덱스]
                오른쪽_인덱스 = 오른쪽_인덱스 + 1
            병합된_배열의_인덱스 = 병합된_배열의_인덱스 + 1

        # 남은 요소들 병합
        while 왼쪽_인덱스 < 왼쪽_배열의_길이:
            arr[병합된_배열의_인덱스] = 왼쪽_배열[왼쪽_인덱스]
            왼쪽_인덱스 = 왼쪽_인덱스 + 1
            병합된_배열의_인덱스 = 병합된_배열의_인덱스 + 1

        while 오른쪽_인덱스 < 오른쪽_배열의_길이:
            arr[병합된_배열의_인덱스] = 오른쪽_배열[오른쪽_인덱스]
            오른쪽_인덱스 = 오른쪽_인덱스 + 1
            병합된_배열의_인덱스 = 병합된_배열의_인덱스 + 1


```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # 중간을 기준으로 배열을 분할
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # 분할된 배열에 대해 재귀적으로 머지 정렬
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # 정렬된 배열을 병합
    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0

    # 두 배열을 비교하여 정렬하며 병합
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 남은 요소들을 추가
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 예시
arr = [12, 11, 13, 5, 6, 7]
sorted_arr = merge_sort(arr)
print("정렬된 배열:", sorted_arr)
```

### Divide and Conquer 
is a problem-solving strategy that involves breaking down a complex problem into smaller, more manageable subproblems. Each subproblem is solved independently, and then the solutions are combined to solve the original problem. This approach typically follows three steps:<br/><br/>

- [ ] Divide: The main problem is divided into smaller, more manageable subproblems of the same type.

- [ ] Conquer: Each subproblem is solved independently using the same divide and conquer strategy. If the subproblem is small enough, it's solved directly.

- [ ] Combine: The solutions to the subproblems are combined to obtain the solution for the original problem.
<br/><br/>

This strategy is commonly used in algorithms and algorithms design to improve efficiency and solve complex problems by breaking them down into simpler, more manageable parts.<br/><br/><br/><br/>

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    # 배열을 반으로 나누기
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # 각 부분을 재귀적으로 정렬
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # 정렬된 부분을 병합
    return merge(left_half, right_half)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 예시 배열
example_array = [12, 11, 13, 5, 6, 7]

# 병합 정렬을 이용하여 배열 정렬
sorted_array = merge_sort(example_array)
print("정렬된 배열:", sorted_array)
```

## 트리 정보

### 기본 용어

- **노드 (Node)**: 트리의 기본 요소로, 데이터와 자식 노드들에 대한 참조를 가지고 있습니다.

- **루트 노드 (Root Node)**: 트리의 최상위 노드로, 다른 모든 노드는 루트 노드로부터의 경로를 가집니다.

- **부모 노드 (Parent Node)**: 다른 노드와 연결된 노드로, 상위 계층의 노드를 가리킵니다.

- **자식 노드 (Child Node)**: 부모 노드와 연결된 노드로, 하위 계층의 노드를 가리킵니다.

- **리프 노드 (Leaf Node)**: 자식 노드가 없는 노드로, 트리의 끝 부분에 위치한 노드입니다.

- **서브 트리 (Subtree)**: 트리 내에서 특정 노드와 그의 자손들로 구성된 부분 트리입니다.

### 트리의 속성

- **트리 사이즈 (Tree Size)**: 트리에 포함된 전체 노드의 수를 나타냅니다.

- **트리 길이 (Tree Height)**: 트리의 루트 노드에서 가장 깊은 리프 노드까지의 경로의 길이를 나타냅니다.

- **트리 깊이 (Node Depth)**: 각 노드의 깊이는 해당 노드가 루트 노드로부터 얼마나 떨어져 있는지를 나타냅니다.
- **트리의 높이 (Tree Height)** : 트리의 높이는 트리 내에서 가장 깊은 노드까지의 경로의 길이를 나타냅니다. 루트 노드에서 가장 깊은 리프 노드까지의 경로 중 가장 긴 경로의 길이를 의미합니다.

- **트리의 차수 (Tree Degree)** : 트리의 차수는 각 노드가 가지는 자식의 수를 나타냅니다. 차수가 크면 큰 트리일수록 복잡한 구조를 가질 수 있습니다.

## Max-Heapify

    Max-Heapify(배열 A, 노드의 인덱스 i):
    1. left_child = 2 * i + 1
    2. right_child = 2 * i + 2
    3. largest = i
    4. if left_child < 배열 A의 길이 and A[left_child] > A[largest]:
        4.1 largest = left_child
    5. if right_child < 배열 A의 길이 and A[right_child] > A[largest]:
        5.1 largest = right_child
    6. if largest != i:
        6.1 A[i]와 A[largest]를 교환
        6.2 Max-Heapify(A, largest)
        

$$
T(n) \leq T\left(\frac{3}{2}n\right) + \Theta(1) 
$$

```python
def max_heapify(arr, n, i):
    largest = i  # 초기값 설정
    left_child = 2 * i + 1
    right_child = 2 * i + 2

    # 왼쪽 자식이 루트보다 크다면
    if left_child < n and arr[left_child] > arr[largest]:
        largest = left_child

    # 오른쪽 자식이 루트보다 크다면
    if right_child < n and arr[right_child] > arr[largest]:
        largest = right_child

    # 루트가 최대 힙 속성을 만족하지 않는 경우
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # 루트와 가장 큰 자식 교환
        max_heapify(arr, n, largest)  # 재귀적으로 서브트리에 대해 Max-Heapify 호출

# 배열을 최대 힙으로 변환
def build_max_heap(arr):
    n = len(arr)

    # 마지막 non-leaf 노드부터 역순으로 Max-Heapify 호출
    for i in range(n // 2 - 1, -1, -1):
        max_heapify(arr, n, i)

# 예시 사용
arr = [4, 10, 3, 5, 1]
print("원래 배열:", arr)

# 배열을 최대 힙으로 변환
build_max_heap(arr)
print("최대 힙 배열:", arr)
```

## Quick-sort 
<br/><br/>
<div style="border: 2px solid green; padding: 5px; background-color: yellow; border-radius: 5px; color: green; display: inline-block;">
     pseudo-code
</div><br/>
퀵정렬(배열 arr, 정수 시작 인덱스 low, 정수 끝 인덱스 high):
    만약 low < high라면:
        pivot_index = 파티션(arr, low, high)  # 배열을 분할하고 피벗의 위치를 가져옴

        # 피벗을 기준으로 좌우 부분배열을 재귀적으로 정렬
        퀵정렬(arr, low, pivot_index - 1)  # 피벗의 왼쪽 부분배열 정렬
        퀵정렬(arr, pivot_index + 1, high)  # 피벗의 오른쪽 부분배열 정렬

파티션(배열 arr, 정수 low, 정수 high):
    pivot = arr[high]  # 피벗을 배열의 마지막 요소로 선택
    i = low - 1  # 작은 요소들을 위한 인덱스

    for j = low to high - 1:
        # 현재 요소가 피벗보다 작으면 작은 부분배열에 추가
        만약 arr[j] < pivot이면:
            i = i + 1
            arr[i]와 arr[j]를 교환

    arr[i + 1]와 arr[high]를 교환  # 피벗 위치를 변경
    리턴 i + 1  # 피벗의 위치 반환


```python
def partition(arr, low, high):
    # 피벗을 배열의 가장 오른쪽 원소로 선택
    pivot = arr[high]
    i = low - 1  # 작은 요소들을 위한 인덱스

    for j in range(low, high):
        # 현재 요소가 피벗보다 작으면 작은 부분배열에 추가
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    # 피벗의 위치를 변경하여 작은 요소들과 큰 요소들을 분리
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1  # 피벗의 위치 반환

def quick_sort(arr, low, high):
    if low < high:
        # 배열을 파티션하고 피벗의 위치를 가져옴
        pivot_index = partition(arr, low, high)

        # 피벗을 기준으로 좌우 부분배열을 재귀적으로 정렬
        quick_sort(arr, low, pivot_index - 1)  # 피벗의 왼쪽 부분배열 정렬
        quick_sort(arr, pivot_index + 1, high)  # 피벗의 오른쪽 부분배열 정렬

# 퀵 정렬 테스트
arr = [3, 6, 8, 10, 1, 2, 1]
print("원래 배열:", arr)
quick_sort(arr, 0, len(arr) - 1)
print("정렬된 배열:", arr)
```

```python
import random

def partition(arr, low, high):
    # 피벗을 배열에서 무작위로 선택
    random_pivot_index = random.randint(low, high)
    arr[high], arr[random_pivot_index] = arr[random_pivot_index], arr[high]

    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def randomized_quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)

        randomized_quick_sort(arr, low, pivot_index - 1)
        randomized_quick_sort(arr, pivot_index + 1, high)

# 테스트
arr = [3, 6, 8, 10, 1, 2, 1]
print("원래 배열:", arr)

randomized_quick_sort(arr, 0, len(arr) - 1)
print("정렬된 배열:", arr)
```

## Radix-sort 
<br/><br/>

###  Radix-sort  based on a <u>stable</u> sort method
-  it only work when it over condition.<br/>
Ο((𝑛+k))xW(maxium digit in array)
[![YouTube Thumbnail](https://img.youtube.com/vi/7heKwjqkY60/maxresdefault.jpg)](https://www.youtube.com/watch?v=7heKwjqkY60)


```python
from typing import List
import math

def countingSortByDigit(nums:List[int],digit:int)->List[int]:  
  counts = [0]*10
  for num in nums:
    count_idx = num//pow(10,digit)%10
    counts[count_idx] += 1

  acc_counts = []
  acc_count = 0
  for count in counts:
    acc_count += count
    acc_counts.append(acc_count)
  end_locs = [ c-1 for c in acc_counts]

  sorted = [0] * len(nums)
  for num in reversed(nums):
    count_idx = num//pow(10,digit)%10
    end_loc = end_locs[count_idx]
    sorted[end_loc] = num
    end_locs[count_idx] -= 1

  return sorted


def radixSort(nums:List[int])->List[int]:
  largest_num = max(nums)
  digits = int(math.log10(largest_num))+1
  sorted = nums
  for digit in range(digits):
    sorted = countingSortByDigit(nums=sorted,digit=digit)
  return sorted

print(radixSort(nums=[391,582,50,924,8,192]))
```

```python
def bucket_sort(arr):
    n = len(arr)
    # 버킷 생성
    buckets = [[] for _ in range(n)]

    # 각 요소를 버킷에 배정
    for i in range(n):
        index = int(n * arr[i])
        buckets[index].append(arr[i])

    # 각 버킷 정렬
    for i in range(n):
        buckets[i] = sorted(buckets[i])

    # 버킷 결합
    k = 0
    for i in range(n):
        for j in range(len(buckets[i])):
            arr[k] = buckets[i][j]
            k += 1

# 테스트
arr = [0.3, 0.1, 0.7, 0.5, 0.2, 0.6]
print("원래 배열:", arr)
bucket_sort(arr)
print("정렬된 배열:", arr)
```

# The greedy algorithm

```python
def knapsack_greedy(weights, values, capacity):
    n = len(weights)
    # 각 물건의 가치 대 무게 비율을 계산하여 내림차순으로 정렬
    value_per_weight = [(values[i] / weights[i], weights[i], values[i], i) for i in range(n)]
    value_per_weight.sort(reverse=True, key=lambda x: x[0])

    max_value = 0
    selected_items = [0] * n  # 선택된 물건을 나타내는 리스트 (1: 선택, 0: 미선택)

    for _, weight, value, i in value_per_weight:
        if capacity >= weight:
            max_value += value
            selected_items[i] = 1
            capacity -= weight

    return max_value, selected_items

# 예시
weights = [2, 2, 4, 5, 3]
values = [3, 7, 2, 9, 6]
capacity = 10
max_value, selected_items = knapsack_greedy(weights, values, capacity)
print("최대 가치:", max_value)
print("선택된 아이템:", selected_items)
```
