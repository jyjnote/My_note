## **Insertion-sort**
### This algorithm is a basic algorithm. It stores the a[key] in an array starting from [j to n] and changes the position if it is less than a[i].
<br/><br/>

```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        key = arr[i]  # í˜„ì¬ ìš”ì†Œë¥¼ ì„ íƒí•˜ì—¬ ì‚½ì…í•  ìœ„ì¹˜ ì°¾ê¸°
        j = i - 1

        # keyë³´ë‹¤ í° ì›ì†Œë“¤ì„ ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì´ë™
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key  # keyë¥¼ ì •ë ¬ëœ ë¶€ë¶„ì— ì‚½ì…

# ì˜ˆì‹œ
arr = [12, 11, 13, 5, 6]
print("ì •ë ¬ ì „ ë°°ì—´:", arr)

insertion_sort(arr)
print("ì •ë ¬ í›„ ë°°ì—´:", arr)
```

### Divide and Conquer 
is a problem-solving strategy that involves breaking down a complex problem into smaller, more manageable subproblems. Each subproblem is solved independently, and then the solutions are combined to solve the original problem. This approach typically follows three steps:<br/><br/>

- [ ] Divide: The main problem is divided into smaller, more manageable subproblems of the same type.

- [ ] Conquer: Each subproblem is solved independently using the same divide and conquer strategy. If the subproblem is small enough, it's solved directly.

- [ ] Combine: The solutions to the subproblems are combined to obtain the solution for the original problem.
<br/><br/>

This strategy is commonly used in algorithms and algorithms design to improve efficiency and solve complex problems by breaking them down into simpler, more manageable parts.<br/><br/><br/><br/>

## Max-Heapify

    Max-Heapify(ë°°ì—´ A, ë…¸ë“œì˜ ì¸ë±ìŠ¤ i):
    1. left_child = 2 * i + 1
    2. right_child = 2 * i + 2
    3. largest = i
    4. if left_child < ë°°ì—´ Aì˜ ê¸¸ì´ and A[left_child] > A[largest]:
        4.1 largest = left_child
    5. if right_child < ë°°ì—´ Aì˜ ê¸¸ì´ and A[right_child] > A[largest]:
        5.1 largest = right_child
    6. if largest != i:
        6.1 A[i]ì™€ A[largest]ë¥¼ êµí™˜
        6.2 Max-Heapify(A, largest)
        

$$
T(n) \leq T\left(\frac{3}{2}n\right) + \Theta(1) 
$$

```python
def max_heapify(arr, n, i):
    largest = i  # ì´ˆê¸°ê°’ ì„¤ì •
    left_child = 2 * i + 1
    right_child = 2 * i + 2

    # ì™¼ìª½ ìì‹ì´ ë£¨íŠ¸ë³´ë‹¤ í¬ë‹¤ë©´
    if left_child < n and arr[left_child] > arr[largest]:
        largest = left_child

    # ì˜¤ë¥¸ìª½ ìì‹ì´ ë£¨íŠ¸ë³´ë‹¤ í¬ë‹¤ë©´
    if right_child < n and arr[right_child] > arr[largest]:
        largest = right_child

    # ë£¨íŠ¸ê°€ ìµœëŒ€ í™ ì†ì„±ì„ ë§Œì¡±í•˜ì§€ ì•ŠëŠ” ê²½ìš°
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # ë£¨íŠ¸ì™€ ê°€ì¥ í° ìì‹ êµí™˜
        max_heapify(arr, n, largest)  # ì¬ê·€ì ìœ¼ë¡œ ì„œë¸ŒíŠ¸ë¦¬ì— ëŒ€í•´ Max-Heapify í˜¸ì¶œ

# ë°°ì—´ì„ ìµœëŒ€ í™ìœ¼ë¡œ ë³€í™˜
def build_max_heap(arr):
    n = len(arr)

    # ë§ˆì§€ë§‰ non-leaf ë…¸ë“œë¶€í„° ì—­ìˆœìœ¼ë¡œ Max-Heapify í˜¸ì¶œ
    for i in range(n // 2 - 1, -1, -1):
        max_heapify(arr, n, i)

# ì˜ˆì‹œ ì‚¬ìš©
arr = [4, 10, 3, 5, 1]
print("ì›ë˜ ë°°ì—´:", arr)

# ë°°ì—´ì„ ìµœëŒ€ í™ìœ¼ë¡œ ë³€í™˜
build_max_heap(arr)
print("ìµœëŒ€ í™ ë°°ì—´:", arr)
```

## Randomized-quick-sort 
<br/><br/>
Randomized Quick Sort addresses this issue by randomly selecting the pivot element, reducing the probability of encountering worst-case scenarios and ensuring an average time complexity of O(n log n). By randomizing the choice of the pivot, the algorithm is less likely to be influenced by pre-sorted arrays, enhancing its efficiency.
<br/><br/>

  <tr><td>best-case <tr><td>

```python
import random

def partition(arr, low, high):
    # í”¼ë²—ì„ ë°°ì—´ì—ì„œ ë¬´ì‘ìœ„ë¡œ ì„ íƒ
    random_pivot_index = random.randint(low, high)
    arr[high], arr[random_pivot_index] = arr[random_pivot_index], arr[high]

    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def randomized_quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)

        randomized_quick_sort(arr, low, pivot_index - 1)
        randomized_quick_sort(arr, pivot_index + 1, high)

# í…ŒìŠ¤íŠ¸
arr = [3, 6, 8, 10, 1, 2, 1]
print("ì›ë˜ ë°°ì—´:", arr)

randomized_quick_sort(arr, 0, len(arr) - 1)
print("ì •ë ¬ëœ ë°°ì—´:", arr)
```

```python
from typing import List
import math

def countingSortByDigit(nums:List[int],digit:int)->List[int]:  
  counts = [0]*10
  for num in nums:
    count_idx = num//pow(10,digit)%10
    counts[count_idx] += 1

  acc_counts = []
  acc_count = 0
  for count in counts:
    acc_count += count
    acc_counts.append(acc_count)
  end_locs = [ c-1 for c in acc_counts]

  sorted = [0] * len(nums)
  for num in reversed(nums):
    count_idx = num//pow(10,digit)%10
    end_loc = end_locs[count_idx]
    sorted[end_loc] = num
    end_locs[count_idx] -= 1

  return sorted


def radixSort(nums:List[int])->List[int]:
  largest_num = max(nums)
  digits = int(math.log10(largest_num))+1
  sorted = nums
  for digit in range(digits):
    sorted = countingSortByDigit(nums=sorted,digit=digit)
  return sorted

print(radixSort(nums=[391,582,50,924,8,192]))
```

# The greedy algorithm

```python
def knapsack_greedy(weights, values, capacity):
    n = len(weights)
    # ê° ë¬¼ê±´ì˜ ê°€ì¹˜ ëŒ€ ë¬´ê²Œ ë¹„ìœ¨ì„ ê³„ì‚°í•˜ì—¬ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬
    value_per_weight = [(values[i] / weights[i], weights[i], values[i], i) for i in range(n)]
    value_per_weight.sort(reverse=True, key=lambda x: x[0])

    max_value = 0
    selected_items = [0] * n  # ì„ íƒëœ ë¬¼ê±´ì„ ë‚˜íƒ€ë‚´ëŠ” ë¦¬ìŠ¤íŠ¸ (1: ì„ íƒ, 0: ë¯¸ì„ íƒ)

    for _, weight, value, i in value_per_weight:
        if capacity >= weight:
            max_value += value
            selected_items[i] = 1
            capacity -= weight

    return max_value, selected_items

# ì˜ˆì‹œ
weights = [2, 2, 4, 5, 3]
values = [3, 7, 2, 9, 6]
capacity = 10
max_value, selected_items = knapsack_greedy(weights, values, capacity)
print("ìµœëŒ€ ê°€ì¹˜:", max_value)
print("ì„ íƒëœ ì•„ì´í…œ:", selected_items)
```
