## Kafka <sub>https://github.com/CUKykkim/kafka-docker</sub>

*Features:*

**High Throughput and Scalability:**

Kafka can handle a high volume of messages and scale horizontally to accommodate increased load.

**Durability and Fault Tolerance:**

Kafka ensures that messages are durable even in the event of node failures through replication and fault tolerance mechanisms.

**Real-time Data Processing:**

Kafka supports real-time data processing by allowing consumers to subscribe and process messages as soon as they are produced.

**Distributed System:**

Kafka is designed as a distributed system, allowing it to run on a cluster of machines to manage and process large volumes of data.

**Message Retention:**

Kafka allows configurable retention of messages for a specified time, enabling consumers to replay messages or go back in time.

**Pub-Sub Model:**

Kafka follows a publish-subscribe messaging model where producers publish messages to topics, and consumers subscribe to topics to consume messages.

</br>

**Key Concepts:**

**Topic:**

A category or feed name to which messages are published by producers.

**Producer:**

An entity that publishes messages to Kafka topics.

**Consumer:**

An entity that subscribes to Kafka topics and processes the published messages.

**Broker:**

A Kafka server instance that stores and manages the topics, serving producers and consumers.

**Partition:**

A division of a topic that allows Kafka to parallelize message processing and storage.

**Replication:**

The process of duplicating partitions across multiple brokers to ensure fault tolerance and data durability.

**ZooKeeper:**

A distributed coordination service used by Kafka for managing and synchronizing the Kafka brokers.

**Offset:**

A unique identifier assigned to each message within a partition, used to track the progress of consumers.
These concepts and features characterize Kafka as a powerful and versatile distributed streaming platform, capable of handling large-scale, real-time data processing and messaging needs.
