## Big data frame work
- flume

The process of collecting a large volume of logs on a massive server:

Flume: Data Collection Framework
Flume is a distributed framework designed to efficiently collect, aggregate, and move a large amount of log data. It can fetch data from various servers located in different places, configure the flow of data, and deliver integrated data to the destination.
