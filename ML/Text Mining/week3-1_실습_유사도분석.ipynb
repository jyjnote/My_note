```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from konlpy.tag import Komoran

# 데이터 불러오기 (전파누리뉴스 데이터를 CSV 파일 등으로 불러온다고 가정)
data = pd.read_csv(r'C:\Users\kmooc 크롤링\한국방송통신전파진흥원_전파누리_최신뉴스_20230725.csv',
                      encoding='cp949',
                      engine='python')
# 데이터 전처리 (불용어 제거, 특수 문자 제거 등)

# Komoran을 사용하여 명사만 추출
komoran = Komoran()
data['text_nouns'] = data['내용'].apply(lambda x: ' '.join(komoran.nouns(x)))

# Count Vectorize를 사용하여 TDM 생성
vectorizer = CountVectorizer()
tdm = vectorizer.fit_transform(data['text_nouns'])

# TF-IDF Transformer를 사용하여 TF-IDF TDM으로 변환
tfidf_transformer = TfidfTransformer()
tfidf_tdm = tfidf_transformer.fit_transform(tdm)

# 생성한 TDM을 저장 (예: CSV 파일로 저장)
tdm_df = pd.DataFrame(tfidf_tdm.toarray(), columns=vectorizer.get_feature_names_out())
tdm_df.to_csv('tdm.csv', encoding='cp949')
                    
```
