```python
# Importing the libraries
from numpy import loadtxt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
# load the dataset
dataset = loadtxt(r'C:\Users\Desktop\\data\bankruptTrain.csv', delimiter=',')
# split into input (X) and output (y) variables
train_x = dataset[:,0:4]
train_y = dataset[:,4]
print(train_x.shape)
print(train_y.shape)
```

```python
# normalization 
normalizer = preprocessing.Normalization()
normalizer.adapt(np.array(train_x))
print(normalizer.mean.numpy())
```

```python
# define the keras model
model = tf.keras.models.Sequential([
  normalizer,
  tf.keras.layers.Dense(24,  activation='relu'),
  tf.keras.layers.Dense(24, activation='relu'),
  layers.Dropout(0.1),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(train_x, train_y, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(train_x, train_y, verbose=1)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# load test data
dataset_test = loadtxt('bankruptTest.csv', delimiter=',')
# split into input (X) and output (y) variables
test_x = dataset_test[:,0:4]
test_y = dataset_test[:,4]
print(test_x.shape)
print(test_y.shape)
```
