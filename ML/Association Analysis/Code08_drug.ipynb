18가지 legal and illegal drugs 들의 연관성 분석

## 1. 데이터 불러오기 및 전처리

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('drug_consumption.csv',header=None)
cols = ['ID', 'Age','Gender','Edu','Country','Ethnicity','Neuroticism score','Extraversion score',
       'Openness to experience score','Agreeableness score','Conscientiousness score','Impulsiveness score',
        'Sensation seeking score','Alcohol','Amphetamine','Amyl nitrite','Benzodiazepine','Caffeine','Cannabis',
        'Chocolate','Coke','Crack','Ecstacy','Heroin','Ketamine','Legal Highs','LSD','Methadone',
        'Magical Mushroom','Nicotine','Semeron','volatile substance abuse']
data.columns = cols
data
```

```python
data1 = data.drop(['ID', 'Age','Gender','Edu','Country','Ethnicity','Neuroticism score',
  'Extraversion score','Openness to experience score','Agreeableness score','Conscientiousness score',
  'Impulsiveness score','Sensation seeking score','Alcohol','Caffeine','Chocolate','Coke','Nicotine'], axis=1)
data1
```

**해당 약물(식품)을 사용한 적이 없다. (0)**
- CL0 Never Used

**한번이라도 사용했다. (1)**
- CL1 Used over a Decade Ago
- CL2 Used in Last Decade
- CL3 Used in Last Year
- CL4 Used in Last Month
- CL5 Used in Last Week
- CL6 Used in Last Day

```python
data2=pd.DataFrame(np.where(data1=='CL0', 0, 1))
data2.columns = data1.columns
data2
```

## 2. 연관성 분석

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

### 지지도

```python
frequent_itemsets = apriori(data2, min_support=0.1, use_colnames=True, max_len=3)
frequent_itemsets
```

### 신뢰도, 향상도

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5 )
rules = rules.sort_values("lift", ascending=False)
rules.head()
```
