## 1. 데이터 불러오기 및 전처리

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\Desktop\data\movies_user_title.csv")
data['temp']=1
data
```

```python
data[data.duplicated()]
```

```python
data1 = data.pivot(index='userId', columns='title', values='temp')
data1
```

```python
movie_data = data1.fillna(0)
movie_data
```

## 2. 연관성 분석

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

#### apriori 알고리즘 적용

```python
frequent_itemsets = apriori(movie_data, min_support=0.2, use_colnames=True)
frequent_itemsets
```

#### 신뢰도와 향상도 계산

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=3)
rules = rules.sort_values("lift", ascending=False)
rules
```
