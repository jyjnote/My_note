```python
import pandas as pd

dt = pd.read_csv(r"C:\Users\hopio\Desktop\yonsei_practice\data\employees.csv", encoding = 'cp949')
dt.shape
dt
```

```python
dt_listwise = dt.dropna()
dt_listwise.shape
dt_listwise
```

```python
dt_pairwise = dt[~dt["연봉"].isna()]
dt_pairwise.shape
dt_pairwise
```

```python
import scipy.stats as stats
dt_listwise_corr = stats.pearsonr(dt_listwise["입사 연도"], dt_listwise["연봉"])
dt_listwise_corr[1]
```

```python
dt_pairwise_corr = stats.pearsonr(dt_pairwise["입사 연도"], dt_pairwise["연봉"])
dt_pairwise_corr[1]
```
